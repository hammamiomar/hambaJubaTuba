```python
from imagegen import NoiseVisualizer
import numpy as np
from utils import create_mp4_from_pil_images
import os
import torch
```

```python
song="songs/Afrodititita.mp3"
output="dist2BlackForest.mp4"
```

```python
visualizer = NoiseVisualizer()
visualizer.loadSong(song)
```

```python
latents = visualizer.getSpecCircleLatents(distance=2)
fps = visualizer.getFPS()
```

```python
latents.shape
```

```python
basePrompt = "demon dancing cheerfully, blue background, jazzy, bells, creepy and morose, hardcore graphics"
targetPromptChromaScale = ["Black", "Dark Blue", "Purple", "Magenta", "Red", "Orange", "Yellow", "Sea green", 
                           "Green", "Teal", "Light Blue", "White"]

promptEmbeds = visualizer.getPromptEmbeds(basePrompt, targetPromptChromaScale, 0.5)
```

```python
promptEmbeds[0].shape
```

```python
images = visualizer.getVisuals(latents=latents, promptEmbeds=promptEmbeds)
```

```python
create_mp4_from_pil_images(image_array=images,output_path=output,song=song, fps=fps)
```
